```python
from astropy.table import Table
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tilefile = '/Users/forero/tile_01165.fits'
obstargets = Table.read(tilefile, hdu=1)
pottargets = Table.read(tilefile, hdu=2)
skytargets = Table.read(tilefile, hdu=3)
gfatargets = Table.read(tilefile, hdu=4)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(obstargets['RA'], obstargets['DEC'], s=1.0, label='science')
plt.scatter(gfatargets['RA'], gfatargets['DEC'], s=10.0, label='gfa')
plt.scatter(skytargets['RA'], skytargets['DEC'], s=20.0, label='sky')
plt.legend()
```
