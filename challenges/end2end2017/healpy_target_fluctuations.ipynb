```python
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
%matplotlib inline
```

```python
NSIDE_MOCK = 4

n_pixels_mock = hp.nside2npix(NSIDE_MOCK)
print(n_pixels_mock)
pixels_mock = hp.ma(np.arange(n_pixels_mock, dtype=np.double))

pixel_theta, pixel_phi = hp.pix2ang(NSIDE_MOCK, np.arange(n_pixels_mock))

#mask = np.zeros(n_pixels_mock, dtype=np.bool)
#mask[~(pixel_phi > np.pi/2.0) & (pixel_theta<2.0*np.pi/3.0)] = 0
pixels_mock[(pixel_phi > np.pi/2.0) | (pixel_theta>2.0*np.pi/3.0)] = 0
#pixels_mock.mask = mask

hp.mollview(pixels_mock)
```

```python
NSIDE_FLUC = 256

n_pixels_fluc = hp.nside2npix(NSIDE_FLUC)
print(n_pixels_fluc)
pixels_fluc = hp.ma(np.ones(n_pixels_fluc, dtype=np.double))
pixels_fluc = 10*pixels_fluc + np.random.normal(loc=0, scale=1.0, size=n_pixels_fluc)

pixel_theta, pixel_phi = hp.pix2ang(NSIDE_FLUC, np.arange(n_pixels_fluc))

pixels_fluc[(pixel_phi > np.pi/2.0) | (pixel_theta>2.0*np.pi/3.0)] = 0

hp.mollview(pixels_fluc)
```
