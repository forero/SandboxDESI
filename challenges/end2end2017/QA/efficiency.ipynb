```python
import glob
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
bright_name = glob.glob("eff_bgs_bright_epoch*.dat")
faint_name = glob.glob("eff_bgs_faint_epoch*.dat")
```

```python
bright_data = {}
faint_data = {}
for i in range(len(bright_name)):
    bright_data[i] = np.loadtxt(bright_name[i], 
                                dtype=[('z','f8'),('n_in','f8'),('n_out_all','f8'),('n_out_good','f8')])
    faint_data[i] = np.loadtxt(faint_name[i], 
                                dtype=[('z','f8'),('n_in','f8'),('n_out_all','f8'),('n_out_good','f8')])
    
```

```python


fig = plt.figure(1, figsize=(8,6))
plt.rc('font', family='serif', size=20)
epoch = 1
plt.plot(bright_data[epoch]['z'], bright_data[epoch]['n_out_all']/bright_data[epoch]['n_in'], 
         label='BGS bright'.format(epoch), linewidth=3, linestyle='-')
plt.plot(faint_data[epoch]['z'], faint_data[epoch]['n_out_all']/faint_data[epoch]['n_in'], 
         label='BGS faint', linewidth=3, linestyle='--')

plt.xlabel('$z$')
plt.ylabel('$N_{assigned\ to\ fiber}$ / $N_{total}$')
plt.legend(loc=4)
plt.xlim([0,0.60])
plt.ylim([0.9,1.0])
#plt.title("Results after the first 7 months of operation")
plt.savefig("BGS_redshift_efficiency.png", bbox_inches='tight')
```

```python
epoch = 1
plt.plot(bright_data[epoch]['z'], bright_data[epoch]['n_out_good']/bright_data[epoch]['n_in'], 
         label='bright: epoch {}'.format(epoch))
plt.plot(faint_data[epoch]['z'], faint_data[epoch]['n_out_good']/faint_data[epoch]['n_in'], 
         label='faint: epoch {}'.format(epoch))
plt.legend()
```
