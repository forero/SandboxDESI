```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open("fa.log")
lines = f.readlines()
f.close()
```

```python
read_targets = []
read_sky = []
full_calculation = []
full_write = []
for l in lines:
    if "INFO: Read target" in l:
        if "targ-" in l:
            read_targets.append(float(l.split()[-4]))
        if "sky-" in l:
            read_sky.append(float(l.split()[-4]))
    if "INFO: Global timer" in l:
        if "run_assign_full calculation" in l:
            full_calculation.append(float(l.split()[-4]))
        if "run_assign_full write" in l:
            full_write.append(float(l.split()[-4]))
```

```python
#plt.plot(read_targets)
#plt.plot(read_sky)
plt.figure(figsize=(8,5))
plt.rcParams.update({'font.size': 13})

plt.plot(full_calculation, label='FA calculation. Mean {:.1f} s, Min {:.2f} s, Max {:.2f}s'.format(
    np.mean(full_calculation), np.min(full_calculation), np.max(full_calculation)))
plt.plot(full_write, label='Writing. Mean {:.1f} s, Min {:.2f} s, Max {:.2f}s'.format(
    np.mean(full_write), np.min(full_write), np.max(full_write)))
plt.legend()
plt.xlabel("Tile")
plt.ylabel("Time (seconds)")
```
